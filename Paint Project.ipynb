```python
import cv2
import numpy as np

def f1(x):
    pass

# Create white canva
white=np.full((500,500,3),255,dtype=np.uint8)

# Window setup
cv2.namedWindow("canva",cv2.WINDOW_FREERATIO)
cv2.createTrackbar("blue","canva",0,255,f1)
cv2.createTrackbar("green","canva",0,255,f1)
cv2.createTrackbar("red","canva",0,255,f1)
cv2.createTrackbar("thickness","canva",1,25,f1)


# cv2.putText(white,"Press c-circle, r-rect, l-line, f-free draw, e-erase, d-delete all, q-quit",(5,490),cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,0,255),1)

# Function to get color and thickness
def f2():
    bl=cv2.getTrackbarPos("blue","canva")
    gr=cv2.getTrackbarPos("green","canva")
    re=cv2.getTrackbarPos("red","canva")
    th=cv2.getTrackbarPos("thickness","canva")
    return [(bl,gr,re),th]
    

# Mouse callback function
x1=None
y1=None
drawing=False
shape="free"
def f3(event,x,y,flag,user):
    global x1,y1,drawing,shape
    var=f2()
    if event==1:
        drawing=True
        x1=x
        y1=y
    elif event==0:
        if drawing==True and shape=="free":
            cv2.line(white,(x1,y1),(x,y),var[0],var[1])
            x1=x
            y1=y
        if drawing==True and shape=="eraser":
            cv2.line(white,(x1,y1),(x,y),(255,255,255),var[1])
            x1=x
            y1=y
    elif event==4:
        if drawing==True and shape=="rectangle":
            cv2.rectangle(white,(x1,y1),(x,y),var[0],var[1])
        elif drawing==True and shape=="circle":
            cv2.circle(white,(x1,y1),int(((x-x1)**2+(y-y1)**2)**0.5),var[0],var[1])
        elif drawing==True and shape=="line":
            cv2.line(white,(x1,y1),(x,y),var[0],var[1])
            
        drawing=False
        x1=None
        y1=None

    
# Set callback
cv2.setMouseCallback("canva",f3)


while True:

    cv2.putText(white,"Press c-circle, r-rect, l-line, f-free draw, e-erase, d-delete all, q-quit",(5,490),cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,0,255),1)

    cv2.imshow("canva",white)
    key=cv2.waitKey(1)
    if key==ord("f"):
        shape="free" 
    elif key==ord("r"):
        shape="rectangle"
    elif key==ord("c"):
        shape="circle"
    elif key==ord("l"):
        shape="line"
    elif key==ord("d"):
        white[:]=255
    elif key==ord("e"):
        shape="eraser"
    elif key==ord("q"):
        break
cv2.destroyAllWindows()
```
